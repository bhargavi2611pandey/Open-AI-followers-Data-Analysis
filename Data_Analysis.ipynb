# CLEANING THE DATA

```python
# loading the text file in python
with open("finaldata.txt",encoding ='utf-8') as f:
    data = f.read()
```

```python
# spliting the data in chunks (chunks of diffrent user)
chunks = data.split("\n\n")
chunks = [chunk.strip() for chunk in chunks if chunk.strip() != ""]
```

```python
# cleaning the data and storing it in diffrent chunks of variable .
def parsing_data(chunk): 
    chunk = chunk.strip()  # removing all the necessary spaces from the chunk
    sep_chunk = chunk.split("\n") # again seprating the chunk further from new line 
    if (len(sep_chunk)<3):
       print(chunk)
       return None
    user_name = sep_chunk[0]
    posts = int(sep_chunk[1].split("post")[0].replace(",",""))  # removing "," between and converting it to integer.
    followers = float(sep_chunk[2].split("follower")[0].replace(",","").replace("K","").replace("M",""))  # removing "," ,"K" and "M" and converting it to float value
    if "K" in sep_chunk[2]:
        followers = followers*1000 # for users having followers in thousands
    elif "M" in sep_chunk[2]:
        followers = followers*1000000 # for users having followers in millions
    else:
        followers = int(followers) # converting it to interger
    followings = float(sep_chunk[3].split("following")[0].replace(",",""))  # converting it to interger
    name = sep_chunk[4]
    if (len(sep_chunk)>5): # only if a user has type of page and bio info.
       type_page = sep_chunk[5]
       bio = "\n".join(sep_chunk[6:]) # joining the data from 6th index till last index.
    else:  # if the user does not have 
        type_page = "Unkown"
        bio = ""
    return {"USERNAME": user_name,"POSTS": posts, "FOLLOWERS": followers,"FOLLOWINGS":followings , "NAME": name , "TYPE OF PAGE": type_page, "BIO": bio}
print(parsing_data(chunks[7]))    
```

```python
# parsing the data
all_parsed = []
for chunk in chunks:
    print(f"The chunks is :\n{chunk}")
    print("\n\n")
    parsed_data = parsing_data(chunk)
    if parsed_data is not None:
       all_parsed.append(parsed_data)  
```

```python
import json # to convert the file into json format
s = json.dumps(all_parsed, indent = 4)
print(s)
```
